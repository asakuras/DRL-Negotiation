```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.figure(figsize=(12,7))
matplotlib.style.use("seaborn-white")
matplotlib.rc('xtick', labelsize=22) 
matplotlib.rc('ytick', labelsize=22)
c = 1.0
x_space = np.linspace(0,1,100)
plt.plot(x_space, 1 - (x_space)**(1/c), label = "Linear (c=1.0)")

c = 0.3
plt.plot(x_space, 1 - (x_space)**(1/c), label = "Boulware (c=0.3)")

c = 4.0
plt.plot(x_space, 1 - (x_space)**(1/c), label = "Conceder (c=3.0)")
plt.legend(fontsize = 20)
plt.xlabel("Scaled Time", fontsize = 20)
plt.ylabel("Scaled Decision Utility", fontsize = 20)
plt.savefig("time-based-agents.png")
```
